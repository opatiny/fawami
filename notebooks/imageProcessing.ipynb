# Processing an SVG pattern

- the pattern contains all parts in 1 SVG file
- find a way to load the SVG
- split the pattern into individual parts
- only keep the contour of the parts (we won't deal with holes)

## Load SVG file

```python
import os

relPath = "../data/freesewing-aaron.svg"
image = os.path.abspath(os.path.join(os.getcwd(), relPath))
print(image)

from svgpathtools import svg2paths

paths, attributes = svg2paths(image)
print(paths)  # List of path objects
```

## Display an SVG

```python
from IPython.core.display import SVG
SVG(filename='../data/freesewing-aaron.svg')

from IPython.display import SVG, display, HTML
import base64
_html_template='<img width="{}" src="data:image/svg+xml;base64,{}" >'

def svg_to_fixed_width_html_image(svg, width="100%"):
    text = _html_template.format(width, base64.b64encode(svg))
    return HTML(text)

svg_to_fixed_width_html_image(paths)

import cv2
```
